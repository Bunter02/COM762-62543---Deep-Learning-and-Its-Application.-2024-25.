```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Create a range of values from -5 to 5 with a step size of 0.1
x_values = np.arange(-5, 5, 0.1)

# Apply the sigmoid function to each value in the range
y_values = sigmoid(x_values)

# Print the results
for x, y in zip(x_values, y_values):
    print(f"sigmoid({x}) = {y}")
```

```python
# Optionally, plot the results
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Function')
plt.grid(True)
plt.show()
```
